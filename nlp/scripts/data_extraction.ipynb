### Main Function

```python
def get_metadata(path: str) -> dict:
    text = ""
    metadata = {}
    
    with open(path, 'rb') as file:
        pdf = fitz.open(path)
        info = pdf.metadata
        number_of_pages = pdf.pageCount
        if info['title'] == None or info['title'] == 'untitled':
            metadata['title'] = get_title_from_file(path)
        else:
            metadata['title'] = info['title']
    
    metadata['author'] = info['author']
    metadata['producer'] = info['producer']
    metadata['numpages'] = number_of_pages
    #metadata['summary'] = generate_summary(text)
    return metadata
```

```python
def convert_to_html(doc):
    html = doc.getText('html').replace('\n', '')
    text = ""
    
    # Convert doc to HTML    
    soup = BeautifulSoup(html)

    # Extract all 'p' tags
    for element in soup.find_all('p'):
        text += element.get_text()
    return text
```
